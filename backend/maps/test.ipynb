```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from map_function import plot_choropleth
```

```python
csv_path = "./data/csv/data_model.csv"
shapefile_path = "./data/shp/Limite des région 2018.geojson"

df = pd.read_csv(csv_path)
gdf = gpd.read_file(shapefile_path)

# Join data from dataframe B with dataframe A
merged = pd.merge(df, gdf, left_on='region', right_on='Name', how='inner')

fig = plot_choropleth(
    geodf=merged,
    column_to_plot='data',
    label_column='Name',
    label_title='Valeur aléatoire',
    title='Carte des régions selon la donnée',
    cmap='viridis',
    size=1.5,
    text_outline_color='white',
    text_outline_width=2
)
plt.show()
```
